```python
# 1-мерный массив из 100 значений в диапазоне от -10 до 10
```

```python
x_arr1 = np.linspace(-10, 10, 100)
```

```python
# несколько графиков в одном окне

# функции гиперболического тангенса
y_0 = np.tanh(x_arr1)

# функция синуса
y_1 = np.sin(x_arr1)

# функция сигмоиды
def sigmoid(arr):
    return 1 / (1 + np.exp(-arr)) # формула сигмоиды

y_2 = sigmoid(x_arr1)

plt.figure(figsize=(10, 8))


plt.plot(x_arr1, y_0)
plt.plot(x_arr1, y_1)
plt.plot(x_arr1, y_2)

plt.legend(['tanh', 'sin', 'sigmoid'])

plt.show()
```
